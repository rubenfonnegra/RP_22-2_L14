```
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Taking the file - j.png (grayscale)
img = cv2.imread("/content/drive/MyDrive/Docencia/Patrones/2022-2/L14/patrones/Clase4/j.png", 0)
plt.imshow(img, cmap = "gray")
```

## Erosion

```
# Erosion
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)

plt.imshow(erosion, cmap="gray")
```

## Dilataci√≥n

```
# Dilatacion
kernel = np.ones(( 5 , 5 ),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)

plt.imshow(dilation, cmap="gray")
```

## Gradientes morfol√≥gicos

```
# Gradientes morfologicos
kernel = np.ones(( 3 , 3 ),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.imshow(gradient, cmap="gray")
```

## Apertura

```
# Taking the file - j2.png (grayscale)
img = cv2.imread("/content/drive/MyDrive/Docencia/Patrones/2022-2/L14/patrones/Clase4/j2.png", 0)

# Apertura
kernel = np.ones(( 5 , 5 ),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

_ , axes = plt.subplots(1,2, figsize=(6,6))
axes[0].imshow(img, cmap='gray'); axes[0].set_xticks([]); axes[0].set_yticks([])
axes[1].imshow(opening, cmap='gray'); axes[1].set_xticks([]); axes[1].set_yticks([])
```

## Cierre

```
# Taking the file - j3.png (grayscale)
img = cv2.imread("/content/drive/MyDrive/Docencia/Patrones/2022-2/L14/patrones/Clase4/j3.png", 0)

# Cierre
kernel = np.ones(( 5 , 5 ),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

_ , axes = plt.subplots(1,2, figsize=(6,6))
axes[0].imshow(img, cmap='gray'); axes[0].set_xticks([]); axes[0].set_yticks([])
axes[1].imshow(closing, cmap='gray'); axes[1].set_xticks([]); axes[1].set_yticks([])
```
